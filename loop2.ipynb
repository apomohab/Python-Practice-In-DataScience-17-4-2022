### تقوم الطريقة بإنشاء كائن مكرر من الداتافريم ، مما يسمح لنا بتكرار كل عمود من أعمدة  
# itteritems
  .

```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv')

df.head(5)
```

```python
for (ColumnName , ColumData) in df.iteritems():
    print('columnName: ', ColumnName)
    print('ColumnData:' , ColumData.values)
```

```python
#Select Columns

for column in df[["director" , "country"]]:
    
    columnCont = df[column]
    
    print(" \n Column Name : " , column + "\n")
    print(" Column Contents : " , columnCont.values)
```
