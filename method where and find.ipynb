```python
import pandas as pd

df = pd.read_csv("Data_for_Missing_Values.csv")

df.head(5)
```

# Where 

 يتم استخدامها للتحقق من إطار بيانات لشرط واحد أو أكثر وإرجاع النتيجة وفقًا لذلك.


```python
#First one - sorting data 

df.sort_values("Country", inplace = True)

# filter data by most director
filter1 = df["Country"] == "France"

# Filtering data
df.where(filter1, inplace = True)

df
```

```python
# Filter data by 2 rows

#First one - sorting data 

df.sort_values("Country", inplace = True)

# filter data by most director
filter1 = df["Country"] == "France"


#Select tow column
filter2 = df["Age"] > 10 

# Filtering data
df.where(filter1 & filter2 ,  inplace = True)

df
```

```python
df = pd.read_csv("nba.csv")

df.head(5)
```

```python
# dropping null value columns to avoid errors
df.dropna(inplace = True)
 
```

```python
# Search about name has ("S") -
```

```python
#Substring to be searched -البحث عن اللاعب في اسمه حرف  a

sub = 'a'

#Create new column has index (a) in string and showing -1 if not get

df['indexes']  = df['Name'].str.find(sub)

#Show
df
```

```python
#Search about tow character , and step 2 characters in search start from 3.

#Substring to be searched 
sub = 'er'

#start
start = 2

#crearing and passing series to new column
df['indexes'] = df['Name'].str.find(sub, start)

#display
df
```
